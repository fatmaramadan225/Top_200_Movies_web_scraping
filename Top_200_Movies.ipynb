```python
from selenium.webdriver.common.by import By
```

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW")
```

```python
All_Movies= driver.find_elements('xpath',"//td")
```

```python
for movi in All_Movies:
     print(movi.text)
```

```python
data = []
```

```python
for movi in All_Movies:
        st=movi.text
        data.append(st.split(' '))
```

```python
df = pd.DataFrame(data)
```

```python
df[0]
```

```python
df[0].values.reshape(-1,4)
```

```python
new_data=pd.DataFrame(df[0].values.reshape(-1,4),columns=['Rank','Title','Lifetime Gross','Year'])
```

```python
top_200_movies=new_data[:200]
```
